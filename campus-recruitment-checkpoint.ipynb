### Categorical Features

```python
def ctg_dist(df, cols, hue=None,rows=3, columns=3):
    fig, axes = plt.subplots(rows, columns, figsize=(16, 12))
    axes = axes.flatten()

    for i, j in zip(df[cols].columns, axes):
        sns.countplot(x=i,
                      data=df,
                      hue=hue,
                      ax=j,
                      order=df[i].value_counts().index)
        j.tick_params(labelrotation=10)
        total = float(len(df[i]))
        j.set_title(f'{str(i).capitalize()} Distribution')
        
        for p in j.patches:
            height = p.get_height()
            j.text(p.get_x() + p.get_width() / 2.,
                    height + 2,
                    '{:1.2f}%'.format((height / total) * 100),
                    ha='center')
            
        plt.tight_layout()
```

#### Text Value

```python
def convert_text_to_encoded_value(df:pd.DataFrame, categorical_cols):
    
    '''Text => Numeric => One Hot Encoding'''
    
    encoder = OneHotEncoder()

    df_cols = df.columns.tolist()
    for col in categorical_cols:
        if col in df_cols:
            if df[col].dtypes == 'object':
                # Text => Numeric
                encoded_col, encoded_cat = df[col].factorize()
                # Numeric => One Hot Encoding
                one_hot_encoded_col = encoder.fit_transform(encoded_col.reshape(-1, 1))
                print(one_hot_encoded_col.toarray())
                df[col] = one_hot_encoded_col.toarray()
```

```python
convert_text_to_encoded_value(train_x, categorical_cols)
train_x.head(5)
```
